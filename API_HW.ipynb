```python
import json
import pandas as pd
import json
import requests
```

```python
#get stock from the user   MSFT  ORCL AMZN NFLX M TSLA XOM
stocks=input()
```

```python
#TOLD TO PUT MYKEY INSTEAD OF ACTUAL KEY
apikey="MYKEY"

url = "https://yfapi.net/v6/finance/quote"
url_trending = "https://yfapi.net/v1/finance/trending/US"

headers = {
    'x-api-key': apikey
    }

trending = requests.request("GET", url_trending, headers=headers)
trending_json = trending.json()
trending_data = trending_json['finance']['result'][0]['quotes'][:5]

querystring = {"symbols":stocks}

headers = {
    'x-api-key': apikey
    }
response = requests.request("GET", url, headers=headers, params=querystring)
stockjson = response.json()
print(response.text)
```

```python
#print Ticker Name, Full Name of the Stock, Current Market Price, Target Mean Price, 52 week high and 52 week low.
stock_df = pd.DataFrame(columns=['Ticker Name', 'Full Name', 'Current Market Price', 'Target Mean Price', '52 week high', '52 week low']) #SETTING UP DATA FRAME
for stock in stockjson['quoteResponse']['result']:
  if "regularMarketPrice" in stock: #CHECKING TO SEE IF STOCK EXISTS
   mean_price = (stock["regularMarketDayHigh"] + stock["regularMarketDayLow"]) / 2 # CALCULATING MEAN

   stock_data = {'Ticker Name': stock["symbol"],'Full Name': stock["longName"],'Current Market Price': str(stock["regularMarketPrice"]),'Target Mean Price': str(mean_price), '52 week high': str(stock["fiftyTwoWeekHigh"]),'52 week low': str(stock["fiftyTwoWeekLow"])} # DATA FOR CURRENT ITERATION
   stock_df = pd.concat([stock_df, pd.DataFrame([stock_data])], ignore_index=True) #ADDING TO DF

   print("Ticker Name: "+stock["symbol"] + " Full Name"+stock["longName"] + " Price:$" + str(stock["regularMarketPrice"])+" Target Mean Price:$" + str(mean_price)+" 52 week high:$"+str(stock["fiftyTwoWeekHigh"]) + " 52 week low:$"+str(stock["fiftyTwoWeekLow"])) #DISPLAY BACK TO USER

for i in trending_data: #DISPLAYING TRENDING DATA
  print(i["symbol"])

stock_df.to_csv("stocks.csv", index=False) #CONVERTING TO CSV FILE
stock_df.head() # FANCY DISPLAY
```
